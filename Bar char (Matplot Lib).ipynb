```python
## HISTOGRAMS ---> NEED ONLY ONE DATA/TABLE

blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]

plt.hist(blood_sugar, bins = [80,100,125,150], rwidth = 0.5, color = 'g')

#rwidth is the width of the graph with respect to the bins 
```

```python
# To print two types of histogram side by side

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.xlabel('sugar range')
plt.ylabel('Total no. of patients')
plt.title('Blood sugar analysis')

plt.hist([blood_sugar_men,blood_sugar_women], bins = [80,100,125,150], rwidth = 0.95, 
         color = ['green', 'red'], label = ['men', 'women'], orientation ='horizontal')

plt.legend()
```
